```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

font1 = {'family': 'Times New Roman', 'size': 16}

def read_embedding(file_path):
    embedding = []
    print(file_path)
    with open(file_path,'r') as f:
        flag = 0
        for line in f.readlines():
            if flag == 0:
                flag = 1
                continue
            else:
                l_s = line.strip().split("\t")
                id = int(l_s[0])
                embs = [float(i) for i in l_s[1:]]
                embedding.append(embs)

    return np.array(embedding)
```

```python
aggregators = ['kegg_sum', 'kegg_mean', 'kegg_concat', 'kegg']
embeddings_2 = {}
embeddings_3 = {}
embeddings_4 = {}


for a in aggregators:
    file_path = "kegg/drug_embeddings_" + str(a) + ".txt"
    embeddings_2[a] = TSNE(n_components=2,init="pca",random_state=2023).fit_transform(read_embedding(file_path))

    file_path = "kegg/3_drug_embeddings_" + str(a) + ".txt"
    embeddings_3[a] = TSNE(n_components=2,init="pca",random_state=2023).fit_transform(read_embedding(file_path))

    file_path = "kegg/4_drug_embeddings_" + str(a) + ".txt"
    embeddings_4[a] = TSNE(n_components=2,init="pca",random_state=2023).fit_transform(read_embedding(file_path))

font1 = {'family': 'Times New Roman', 'size': 20}
plt.figure(figsize=(14,10),dpi=600)
sum_2 = plt.subplot(4,3,1)
sum_3 = plt.subplot(4,3,2)
sum_4 = plt.subplot(4,3,3)
mean_2 = plt.subplot(4,3,4)
mean_3 = plt.subplot(4,3,5)
mean_4 = plt.subplot(4,3,6)
con_2 = plt.subplot(4,3,7)
con_3 = plt.subplot(4,3,8)
con_4 = plt.subplot(4,3,9)
cap_2 = plt.subplot(4,3,10)
cap_3 = plt.subplot(4,3,11)
cap_4 = plt.subplot(4,3,12)

sum_2.scatter(embeddings_2['kegg_sum'][:,0], embeddings_2['kegg_sum'][:,1], s=1, color='#104f8d')
sum_2.set_title("Layer 2", font1)
sum_2.set_ylabel("Sum",font1)
sum_2.set_ylim(-100,100)
sum_2.set_xlim(-80,80)

sum_3.scatter(embeddings_3['kegg_sum'][:,0], embeddings_3['kegg_sum'][:,1], s=1, color='#b04f48')
sum_3.set_title("Layer 3", font1)
sum_3.set_ylim(-100,100)
sum_3.set_xlim(-80,80)

sum_4.scatter(embeddings_4['kegg_sum'][:,0], embeddings_4['kegg_sum'][:,1], s=1, color='#edcc86')
sum_4.set_title("Layer 4", font1)
sum_4.set_ylim(-100,100)
sum_4.set_xlim(-80,80)

mean_2.scatter(embeddings_2['kegg_mean'][:,0], embeddings_2['kegg_mean'][:,1], s=1, color='#104f8d')
mean_2.set_ylabel("Mean",font1)
mean_2.set_ylim(-100, 100)
mean_2.set_xlim(-80,80)
mean_3.scatter(embeddings_3['kegg_mean'][:,0], embeddings_3['kegg_mean'][:,1], s=1, color='#b04f48')
mean_3.set_ylim(-100, 100)
mean_3.set_xlim(-80,80)
mean_4.scatter(embeddings_4['kegg_mean'][:,0], embeddings_4['kegg_mean'][:,1], s=1, color='#edcc86')
mean_4.set_ylim(-100, 100)
mean_4.set_xlim(-80,80)

con_2.scatter(embeddings_2['kegg_concat'][:,0], embeddings_2['kegg_concat'][:,1], s=1, color='#104f8d')
con_2.set_ylabel("Concat",font1)
con_2.set_ylim(-100, 100)
con_2.set_xlim(-80,80)
con_3.scatter(embeddings_3['kegg_concat'][:,0], embeddings_3['kegg_concat'][:,1], s=1, color='#b04f48')
con_3.set_ylim(-100, 100)
con_3.set_xlim(-80,80)
con_4.scatter(embeddings_4['kegg_concat'][:,0], embeddings_4['kegg_concat'][:,1], s=1, color='#edcc86')
con_4.set_ylim(-100, 100)
con_4.set_xlim(-80,80)

cap_2.scatter(embeddings_2['kegg'][:,0], embeddings_2['kegg'][:,1], s=1, color='#104f8d')
cap_2.set_ylabel("Capsule",font1)
cap_2.set_ylim(-100, 100)
cap_2.set_xlim(-80,80)
cap_3.scatter(embeddings_3['kegg'][:,0], embeddings_3['kegg'][:,1], s=1, color='#b04f48')
cap_4.scatter(embeddings_4['kegg'][:,0], embeddings_4['kegg'][:,1], s=1, color='#edcc86')
cap_3.set_ylim(-100, 100)
cap_3.set_xlim(-80,80)
cap_4.set_ylim(-100, 100)
cap_4.set_xlim(-80,80)

plt.tight_layout()
#plt.savefig('drug_embeddings_kegg.svg')
#plt.savefig('drug_embeddings_kegg.jpg')
```
